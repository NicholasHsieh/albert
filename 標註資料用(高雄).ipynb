```python
import os, json 
import pandas as pd 

path_to_json = 'D:\判決書資料\標註資料集(高雄)' 
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')] 
print(json_files[0])
```

```python
with open('D:/判決書資料/標註資料集(高雄)/'+json_files[0],"r",encoding="utf-8") as f:
    data = json.load(f)
```

```python
dic = {
    "court": [data["court"]], 
    "date": [data["date"]],
    "no": [data["no"]],
    "sys": [data["sys"]],
    "reason": [data["reason"]],
    "type": [data["type"]],
    "mainText": [data["mainText"]]
}
df = pd.DataFrame(dic)
df
```

```python
for x in range(1,42836):
    with open('D:/判決書資料/標註資料集(高雄)/'+json_files[x],"r",encoding="utf-8") as f:
        data = json.load(f)
    
    try:
        new = pd.DataFrame({
        "court": [data["court"]], 
        "date": [data["date"]],
        "no": [data["no"]],
        "sys": [data["sys"]],
        "reason": [data["reason"]],
        "type": [data["type"]],
        "mainText": [data["mainText"]]  
        })
        #print(x)
        
    except:
        print(x)

    df = df.append(new)
```

```python
df.to_csv('D:/Result.csv') 
```
